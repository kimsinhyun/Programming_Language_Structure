```python
import pandas as pd
import os
path = "./log_data"
file_lst = os.listdir(path)

# 현재 디렉토리내에 각각의 파일을 출력
ID_list = []
for file in file_lst:
    filepath = path + '/' + file
    ID_list.append(file.split('_')[1])
for i in range(len(ID_list)):
    try:
        try:
            data = pd.read_csv('./log_data/hs_{}_m08_0903_1355.csv'.format(ID_list[i]),engine='python', encoding= 'cp949')
        except:
            data = pd.read_csv('./log_data/hs_{}_m08_0903_1356.csv'.format(ID_list[i]),engine='python', encoding= 'cp949')
    except:
        try:
            data = pd.read_csv('./log_data/hs_{}_m08_0903_1355.csv'.format(ID_list[i]))
        except:
            data = pd.read_csv('./log_data/hs_{}_m08_0903_1356.csv'.format(ID_list[i]))
    data.to_csv('./log_data/hs_{}_m08_0903_1355.csv'.format(ID_list[i]),encoding='utf-8-sig')
```
