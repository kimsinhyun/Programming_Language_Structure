## 부동, 활동 비율

```python
data
```

```python
data['Z'].value_counts()
```

```python
Z_value_count = data['Z'].value_counts()

활동 = 0
부동 = 0
미동 = 0
매우활동=0
total_sum = 0
if "활동" in Z_value_count:
    total_sum += Z_value_count["활동"]
    활동 = Z_value_count["활동"]
if "부동" in Z_value_count:
    total_sum += Z_value_count["부동"]
    부동 = Z_value_count["부동"]
if "미동" in Z_value_count:
    total_sum += Z_value_count["미동"]
    미동 = Z_value_count["미동"]
if "매우 활동" in Z_value_count:
    total_sum += Z_value_count["매우 활동"]
    매우활동 = Z_value_count["매우 활동"]
arr = [[부동/total_sum*100,미동/total_sum*100,활동/total_sum*100,매우활동/total_sum*100]]

df_summary = pd.DataFrame(arr,columns = ["부동","미동","활동", "매우 활동"], index = [""])
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
ax = df_summary.plot(kind='barh',title="활동, 부동 비율", rot = 0, stacked=True, width = 0.1)
for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    if width>0:
      ax.annotate("%.1f"%(width)+"%", xy=(left+width/2, bottom+height/2), ha='center', va='center')
    
plt.sca(ax)
plt.box(False)
plt.show()
```
